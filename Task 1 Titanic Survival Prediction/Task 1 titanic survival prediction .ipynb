# Task 1 Titanic Survival Prediction

## •Model used: Logistic Regression from Scikit-learn. This model was used to predict the survival of passengers on the Titanic based on features like ‘Pclass’ and ‘Sex’..


## •Data Preprocessing :the dataset was transformed from categorical to numerical data using LabelEncoder from Scikit-learn. The data was then split into training and testing sets using train_test_split from Scikit-learn.

## •Visualizations: Various aspects of the data were visualized using Seaborn and Matplotlib. These visualizations included count plots and bar plots of different columns like ‘Survived’, ‘Pclass’.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=df['Sex'], y=df['Survived'], hue=df['Sex'], palette='viridis')
plt.legend([],[], frameon=False)
plt.title('Survival Rate Across Genders', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Survival Rate', fontsize=14)
plt.xticks(ticks=[0,1], labels=['Male', 'Female'])
plt.show()
```

```python
labelencoder = LabelEncoder()
df['Sex'] = labelencoder.fit_transform(df['Sex'])
df.head(15)
```

```python
df=df.drop(['Age'], axis=1)
df_final = df
df_final.head(15)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X = df[['Pclass', 'Sex']]
Y = df['Survived']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)

log = LogisticRegression(random_state = 0)
log.fit(X_train, Y_train)
```

```python
pred = print(log.predict(X_test))
```
