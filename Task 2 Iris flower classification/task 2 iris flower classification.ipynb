Performing exploratory data analysis (EDA) on the Iris dataset. EDA is a way to understand the characteristics of the data by summarizing its main characteristics often with visual methods.

A brief description of the steps taken:

1.Data Loading: Loaded the Iris dataset using pandas, which is a popular library for data manipulation and analysis.

2.Data Exploration: Explored the data by viewing the first few rows, generating descriptive statistics, and counting the number of samples in each class.

3.Data Visualization: Visualized the data using histograms and scatter plots. Histograms provide a visual interpretation of numerical data by indicating the number of data points that lie within a range of values, known as a bin. Scatter plots display values for two variables from your dataset.

4.Correlation Analysis: Computed the correlation matrix of your data and visualized it using a heatmap. Correlation is a statistical measure that expresses the extent to which two variables are linearly related. A heatmap is a graphical representation where individual values of a matrix are represented as colors.

5.Boxplot Generation: Created boxplots for each feature in your dataset. A boxplot is a standardized way of displaying the dataset based on a five-number summary: the minimum, the maximum, the sample median, and the first and third quartiles.

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading the CSV file
df = pd.read_csv('Iris.csv', index_col=0)
df.head()
```

```python
# Creating a histogram for 'sepal_length'
# 'bins' parameter defines the number of equal-width bins in the range
# 'color' parameter sets the color of the histogram
# 'edgecolor' parameter sets the color of the bin edges
pd.Series(df.index).hist(bins=10, color='lightblue', edgecolor='black')

# Adding title and labels
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')

# Displaying the plot
plt.show()
```

```python
# Creating a histogram for 'sepal_width'
# 'bins' parameter defines the number of equal-width bins in the range
# 'color' parameter sets the color of the histogram
# 'edgecolor' parameter sets the color of the bin edges
df['sepal_width'].hist(bins=10, color='lightblue', edgecolor='black')

# Adding title and labels
plt.title('Histogram of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')

# Displaying the plot
plt.show()
```

```python
# Creating a histogram for 'petal_length'
# 'bins' parameter defines the number of equal-width bins in the range
# 'color' parameter sets the color of the histogram
# 'edgecolor' parameter sets the color of the bin edges
df['petal_length'].hist(bins=10, color='lightblue', edgecolor='black')

# Adding title and labels
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')

# Displaying the plot
plt.show()
```

```python
# Creating a histogram for 'petal_width'
# 'bins' parameter defines the number of equal-width bins in the range
# 'color' parameter sets the color of the histogram
# 'edgecolor' parameter sets the color of the bin edges
df['petal_width'].hist(bins=10, color='lightblue', edgecolor='black')

# Adding title and labels
plt.title('Histogram of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')

# Displaying the plot
plt.show()
```

```python
# Defining colors and species
colors = ['red', 'orange', 'blue']
species = ['Iris-virginica','Iris-versicolor','Iris-setosa']

# Creating a scatter plot for each species
for i in range(3):
    # Filtering data for the current species
    x = df[df['species'] == species[i]]
    
    # Creating a scatter plot with sepal length on the x-axis and sepal width on the y-axis
    # 'c' parameter sets the color of the points
    # 'label' parameter sets the label for the points, which will be used in the legend
    plt.scatter(x.index, x['sepal_width'], c = colors[i], label=species[i])

# Adding title and labels
plt.title('Scatter Plot of Sepal Dimensions')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Adding a legend
plt.legend()

# Displaying the plot
plt.show()
```

```python
# Creating a boxplot for each feature
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for feature in features:
    sns.boxplot(x='species', y=feature, data=df)
    plt.title('Boxplot of ' + feature.capitalize())
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['species'] = le.fit_transform(df['species'])
df.head()
```

```python
#Calculating the correlation matrix
corr = df.corr()

# Creating a new figure and axes with specified figure size
fig, ax = plt.subplots(figsize=(8,6))

# Creating a heatmap of the correlation matrix
# 'annot' parameter is set to True to write the data value in each cell
# 'ax' parameter determines the axes of the plot
# 'cmap' parameter sets the colormap of the heatmap
sns.heatmap(corr, annot=True, ax=ax, cmap='coolwarm', fmt=".2f", linewidths=.05)

# Adding title
plt.title('Heatmap of Correlation Matrix')

# Displaying the plot
plt.show()
```

```python
from sklearn.model_selection import train_test_split
# train - 70
# test - 30
X = df.drop(columns=['species'])
Y = df['species']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
# logistic regression 
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
# model training
model.fit(x_train, y_train)
```

```python
# print metric to get performance
print("Accuracy: ",model.score(x_test, y_test) * 100)
```
